```python
nomes = ['Danilo','Alberto','Lucelia','Jhonattan','Gabriel','Rafael','Luiz','Lucas_Pinho','Victor','Ulisses','Mauro','Lucas_Silva']
idades = [34,38,40,24,27,29,20,32,33,45,23,35]

nomes_idades = []

for n,i in zip(nomes,idades):
  nomes_idades.append([n,i])

titulo = ['Nomes','Idades']

nomes_idades.insert(0,titulo)
nomes_idades
```

Etapa 1 b)

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize = (18,10))
plt.bar(nomes,idades)
```

Etapa 2

```python
salario = []

for indice,sal in enumerate(nomes):
  salario.append(float(input(f"Qual salário você quer receber, {nomes[indice]}? ")))
  print(salario)
  
```

```python

nomes_idades_salario = []

for n,i,s in zip(nomes,idades, salario):
  nomes_idades_salario.append([n,i,s])

titulo = ['Nomes','Idades','Salario']

nomes_idades_salario.insert(0,titulo)
nomes_idades_salario
```

Etapa 2: gráfico

```python
import numpy as np
import matplotlib.pyplot as plt

plt.figure(figsize = (18,10))
plt.bar(nomes,salario)
```
